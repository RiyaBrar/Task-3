### Data Visualizations

```python
plt.figure(figsize=(20,8))              #
df['State'].value_counts().plot.bar()
plt.xlabel('State',size=14)
plt.xticks(size=10)
plt.ylabel('Number of Orders', size=16)
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(5,5))
colors = ['pink', 'lightblue','orange','lightgreen']

df['Ship Mode'].value_counts().plot.pie(shadow = True,colors=colors,explode=(0,0.1,0.1,0.1) ,autopct = '%.2f%%')
#autopct enables you to display the percent value using Python string formatting. 
plt.title('Ship Mode', fontsize = 20)
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(7,7))
colors = ['pink', 'lightblue','lightgreen']
df['Category'].value_counts().plot.pie(shadow = True,colors=colors,explode=(0,0.1,0.1) ,autopct = '%.2f%%')
#autopct enables you to display the percent value using Python string formatting. 
plt.title('Category', fontsize = 20)
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(9,9))

df.groupby('Sub-Category').sum()['Quantity'].plot.pie(shadow = True ,autopct = '%.2f%%')
#autopct enables you to display the percent value using Python string formatting. 
plt.title('Sub-Category as per Quantity-wise', fontsize = 20)
plt.legend()
plt.show()
```

```python
sales=pd.DataFrame(df.groupby('Sub-Category').sum()['Sales'])
sales
```

```python
# Visualization of the sales on the bar graph
sales.plot.bar(figsize=(10,5))
plt.title('Sales of sub-category items',size=18)
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.scatter("Sub-Category","Profit",c="orange",data=df)
plt.xticks(rotation="vertical", size=10)
plt.xlabel("Sub-Category")
plt.ylabel("profit")
plt.title('Scatter plot of Profits on sub-category items',size=18)
plt.grid()
plt.show()
```

```python
profits=pd.DataFrame(df.groupby("Sub-Category").sum()["Profit"])
profits
```

```python
profits=profits.sort_values('Profit')# sorting profits as sub-category in incresing order
profits
```

```python
profits.plot.bar(figsize=(10,5))
plt.title('graph of profits on sub-category items',size=18)
plt.grid()
plt.show()
```

```python
colours = ['lightgreen', 'pink']
df.groupby('Sub-Category')[['Profit','Sales']].sum().plot.bar(figsize=(10,5),color=colours)
plt.title('Aggregate profits and sales per Sub-Category',size=20)
plt.grid()
plt.show()
```

### Profit per State-wise in decreasing order

```python
state_profit=pd.DataFrame(df.groupby('State').sum()['Profit'])
state_profit.sort_values('Profit',ascending=False)
```

```python
plt.figure(figsize=(20,15))
statewise=df.groupby(['State'])['Profit'].sum().nlargest(50)
statewise.plot.bar()
plt.title('Graph of Statewise profit',size=25)
plt.grid()
plt.show()
```

### The analysis above show that State Texas have the lowest Profit whereas California have maximum profit

```python
plt.figure(figsize=(10,5))
df.groupby('Region').sum()['Profit'].plot.pie(shadow = True ,colors=['pink', 'lightblue','lightgreen','orange'],explode=(0,0,0,0.1),autopct = '%.2f%%')
plt.title('Profit of per-region',size=18)
plt.legend()
plt.show()
```

### This pie chart shows that West Region has the most profit, whereas Central region has the least profit
